#  Question 4 - Extraction des revenus de GameStop avec Web Scraping

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager

# Step 1: Setup the webdriver
driver = webdriver.Chrome(ChromeDriverManager().install())

# Step 2: Open the GameStop revenue page
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"  # Example URL for GME revenue data
driver.get(url)

# Step 3: Locate the table with the revenue data
# Wait until the table is loaded (optional: use WebDriverWait for better control)
table = driver.find_element(By.XPATH, "//table[@class='historical_data_table']")

# Step 4: Extract the table rows
rows = table.find_elements(By.TAG_NAME, "tr")
table_data = []

# Step 5: Loop through each row and extract the data
for row in rows[1:]:  # Skip header row
    cells = row.find_elements(By.TAG_NAME, "td")
    if len(cells) > 0:
        year = cells[0].text.strip()
        revenue = cells[1].text.strip().replace(',', '').replace('$', '')  # Clean up revenue data
        table_data.append([year, revenue])

# Step 6: Create a DataFrame
gme_revenue = pd.DataFrame(table_data, columns=['Year', 'Revenue'])
gme_revenue['Revenue'] = pd.to_numeric(gme_revenue['Revenue'], errors='coerce')

# Step 7: Display the last five rows
print(gme_revenue.tail())

# Close the driver
driver.quit()
```

# Question 5 - Tableau de bord : actions et revenus de Tesla

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(ticker_symbol):
    # Download stock data using yfinance
    stock_data = yf.Ticker(ticker_symbol)
    stock_data = stock_data.history(period="1y")  # Last 1 year of data
    
    # Plot the stock price data
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data["Close"], label=f"{ticker_symbol} Stock Price")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.title(f"{ticker_symbol} Stock Price Over the Last Year")
    plt.legend(loc="upper left")
    plt.grid(True)
    plt.show()

# Use the function for Tesla stock
make_graph("TSLA")
```
